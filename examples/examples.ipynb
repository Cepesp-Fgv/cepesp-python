## Api cepesp data em python
### Utilização básica
Votação para presidente 

```python
from cepesp import *
vot = votos(ano=2014,
            cargo=CARGO.PRESIDENTE,
            agregacao_regional=AGR_REGIONAL.BRASIL,
            agregacao_politica=AGR_POLITICA.CANDIDATO)
vot[["NUMERO_CANDIDATO","NUM_TURNO","QTDE_VOTOS"]].head()
```

####Adicionando filtros

```python
vot = votos(ano=2014,
            cargo=CARGO.PRESIDENTE,
            agregacao_regional=AGR_REGIONAL.UF,
            estado="CE",
            numero_candidato="13")
vot[["NUMERO_CANDIDATO","UF","NUM_TURNO","QTDE_VOTOS"]].head()
```

#### Informações sobre o candidato


```python
cand = candidatos(ano=2014,cargo=CARGO.DEPUTADO_FEDERAL)
cand[["NUMERO_CANDIDATO","NOME_CANDIDATO","SIGLA_UF","NUM_TURNO"]].head()
```

### Informações sobre Legendas


```python
leg = legendas(ano=2014,cargo=CARGO.GOVERNADOR)
leg.head()
```

### Junção entre votação e candidatos

```python
vot = votos_x_candidatos(ano=2014,
                         cargo=CARGO.PRESIDENTE,
                         agregacao_regional=AGR_REGIONAL.BRASIL).head()
vot.head()
```

### Votos x Partidos

```python
votos_x_legendas(ano=2014,
                 cargo=CARGO.PRESIDENTE,
                 agregacao_regional=AGR_REGIONAL.BRASIL).head()
```
